# Day 3 — Trees, Random Forests, Gradient Boosting
Use the Titanic dataset. Compare DT vs RF vs XGBoost vs LightGBM.


````python
import pandas as pd, numpy as np
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# Get Titanic via seaborn for convenience
import seaborn as sns
df = sns.load_dataset('titanic').dropna(subset=['age', 'embarked'])
y = df['survived']
X = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']].copy()
X = pd.get_dummies(X, columns=['sex', 'embarked'], drop_first=True)
````

````python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
models = {
    'DecisionTree': DecisionTreeClassifier(max_depth=5, random_state=0),
    'RandomForest': RandomForestClassifier(n_estimators=300, random_state=0, n_jobs=-1),
    'XGBoost': XGBClassifier(n_estimators=300, max_depth=4, learning_rate=0.05, eval_metric='logloss', random_state=0),
    'LightGBM': LGBMClassifier(n_estimators=300, max_depth=-1, learning_rate=0.05, random_state=0),
}
for name, m in models.items():
    scores = cross_val_score(m, X, y, scoring='f1', cv=cv, n_jobs=-1)
    print(f'{name:14s}  F1 = {scores.mean():.3f} ± {scores.std():.3f}')
````

## Feature importance (XGBoost)

````python
import matplotlib.pyplot as plt
m = models['XGBoost'].fit(X, y)
imp = pd.Series(m.feature_importances_, index=X.columns).sort_values()
imp.plot.barh(); plt.title('XGBoost feature importance'); plt.tight_layout();
````

Train a YOLOv8 object detector on a small custom dataset.

Install: `pip install ultralytics roboflow`

Easiest dataset: pull a small one from Roboflow Universe in YOLO format. For a no-download demo we just run inference with a pretrained model.


## 1. Inference with the pretrained YOLOv8n on COCO classes

````python
from ultralytics import YOLO
import urllib.request, pathlib

url = 'https://ultralytics.com/images/bus.jpg'
img = pathlib.Path('bus.jpg')
if not img.exists():
    urllib.request.urlretrieve(url, img)

model = YOLO('yolov8n.pt')
results = model(str(img))
results[0].show()  # opens annotated image
for box in results[0].boxes:
    cls = int(box.cls.item()); conf = float(box.conf.item())
    print(f'{model.names[cls]:<15s} conf={conf:.2f}  bbox={box.xyxy.tolist()[0]}')
````
